#ðŸ“Œ ExtracciÃ³n

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
js_url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python
df = pd.read_json(js_url)
df
```

```python
pd.set_option('display.max_columns', None)
```

#ðŸ”§ TransformaciÃ³n

```python
df_data = pd.json_normalize(df.to_dict(orient='records'), sep='_')
df_data
```

```python
from os import error
df_data['account_Charges_Total'] = pd.to_numeric(
    df_data['account_Charges_Total'],
    errors='coerce'
)

df_data.info()
```

```python
df_data['account_Charges_Total'].fillna(float(0.0), inplace=True)
df_data['account_Charges_Total'].isna().sum()
```

```python
df_data.info()
```

```python
df_data['Cuentas_Diarias'] = (df_data['account_Charges_Monthly'] / 30).astype(float)
df_data['Cuentas_Diarias'] = df_data['Cuentas_Diarias'].round(2)
df_data
```

```python
# 1. Limpiar nombres
df_data.columns = df_data.columns.str.strip()

# 2. Renombrar
df_data.rename(columns={
    'customer_gender':'gender',
    'customer_SeniorCitizen': 'senior_65',
    'customer_Partner': 'partner',
    'customer_Dependents': 'dependents',
    'customer_tenure': 'tenure_months',
    'phone_PhoneService': 'phone_service',
    'phone_MultipleLines': 'multiple_lines',
    'internet_InternetService':'internet_service',
    'internet_OnlineSecurity': 'online_security',
    'internet_TechSupport': 'tech_support',
    'internet_OnlineBackup': 'online_backup',
    'internet_DeviceProtection': 'device_protection',
    'internet_StreamingTV': 'streaming_tv',
    'internet_StreamingMovies': 'streaming_movies',
    'account_Contract': 'contract_type',
    'account_PaperlessBilling': 'paperless_billing',
    'account_PaymentMethod': 'payment_method',
    'account_Charges_Monthly': 'monthly_charge',
    'account_Charges_Total': 'total_charge',
    'Cuentas_Diarias':'daily_charge'
}, inplace=True)

```

```python
df_data['streaming_tv'].unique()
```

```python
df_data.head()
```

```python
cols_finales = [
    'Churn',
    'senior_65',
    'partner',
    'dependents',
    'tenure_months',
    'contract_type',
    'monthly_charge',
    'daily_charge',
    'total_charge',
    'internet_service',
    'online_security',
    'tech_support',
    'streaming_tv',
    'payment_method'
]

df_final = df_data[cols_finales]
df_final
```

```python
# Reemplazar strings vacÃ­os por NaN
df_final.replace('', np.nan, inplace=True)
```

```python
cols_binarias = [
    'Churn',
    'partner',
    'dependents',
    'online_security',
    'tech_support',
    'streaming_tv'
]
```

```python
map_binario = {'Yes': 1, 'No': 0, 'No internet service':0}

df_final.loc[:, cols_binarias] = (
    df_final[cols_binarias]
    .replace(map_binario)
)
```

```python
df_final[cols_binarias] = (
    df_final[cols_binarias]
    .fillna(0)
    .astype(int)
)
```

```python
df_final['payment_method'].unique()
```

```python
mapa_contrato = {'One year': 0, 'Month-to-month':1,'Two year':2}
mapa_internet = {'DSL':0,'Fiber optic':1,'No':2}
mapa_payment = {'Mailed check':0,'Electronic check':1,'Credit card (automatic)':2,'Bank transfer (automatic)':3}
```

```python
df_final['contract_type'] = df_final['contract_type'].map(mapa_contrato)
df_final['internet_service'] = df_final['internet_service'].map(mapa_internet)
df_final['payment_method'] = df_final['payment_method'].map(mapa_payment)

df_final
```

#ðŸ“Š Carga y anÃ¡lisis

```python
df_final.describe()
```

```python
df_final.groupby('Churn').mean()
```

```python
cols_validas = [
    'senior_65', 'partner', 'dependents',
    'tenure_months','contract_type',
    'monthly_charge', 'daily_charge', 'total_charge',
    'online_security', 'tech_support', 'streaming_tv','payment_method'
]

df_grup = df_final.groupby('Churn')[cols_validas].mean()
df_grup
```

```python
df_grup_reset = df_grup.reset_index()

grouped_melt = df_grup_reset.melt(
    id_vars='Churn',
    value_vars=['tenure_months', 'monthly_charge'],
    var_name='Variable',
    value_name='Promedio'
)

ax = sns.barplot(
    data=grouped_melt,
    x='Variable',
    y='Promedio',
    hue='Churn'
)

handles, labels = ax.get_legend_handles_labels()
ax.legend(
    handles,
    ['No cancelÃ³', 'CancelÃ³'],
    title='Estado de cancelaciÃ³n'
)

plt.title('ComparaciÃ³n de promedios por Churn')
plt.xticks([0, 1], ['Meses Contratados', 'Cargo mensual'])
plt.show()



```

```python
df_contratos = df_final.groupby(['Churn', 'contract_type']).size().reset_index(name='count')

df_contratos['total'] = df_contratos.groupby('contract_type')['count'].transform('sum')
df_contratos['churn_rate'] = df_contratos['count'] / df_contratos['total']

df_contratos
```

```python
sns.barplot(
    data=df_contratos[df_contratos['Churn'] == 1],
    x='contract_type',
    y='churn_rate'
)

plt.title('Tasa de cancelaciÃ³n por tipo de contrato')
plt.ylabel('ProporciÃ³n de cancelacion')
plt.xlabel('Tipo de contrato')
plt.xticks([0,1,2], ['Anual', 'Mensual', '2 AÃ±os'])
plt.show()
```

```python
df_pagos = df_final.groupby(['Churn', 'payment_method']).size().reset_index(name='count')
df_pagos['total'] = df_pagos.groupby('payment_method')['count'].transform('sum')
df_pagos['churn_rate'] = df_pagos['count'] / df_pagos['total'] * 100

df_pagos
```

```python
sns.barplot(
    data=df_pagos[df_pagos['Churn'] == 1],
    x='payment_method',
    y='churn_rate'
)

plt.title('Tasa de cancelaciÃ³n por mÃ©todo de pago')
plt.ylabel('ProporciÃ³n de cancelacion')
plt.xlabel('MÃ©todo de pago')
plt.xticks([0,1,2,3],['Cheques', 'Cheque Electronico', 'Tarjeta de crÃ©dito', 'Transferencia bancaria'], rotation=10)
plt.show()
```

```python
sns.boxplot(data=df_final, x='Churn', y='monthly_charge')
plt.title('DistribuciÃ³n de cargos mensuales')
plt.xticks([0,1], ['No cancelÃ³', 'CancelÃ³'])
plt.title('DistribuciÃ³n de cancelacion por cargos mensuales')
plt.show()
```

```python
sns.histplot(data=df_final,
             x='monthly_charge',
             hue='Churn',
             kde=True)
plt.title('DistribuciÃ³n de cancelacion por cargos mensuales')
plt.xlabel('Cargo mensual')
plt.ylabel('Cancelaciones')
plt.show()
```
